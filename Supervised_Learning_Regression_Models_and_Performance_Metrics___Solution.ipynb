Q1.  What is Simple Linear Regression (SLR)? Explain its purpose.
>- Simple Linear Regression (SLR) is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.

Purpose:

To predict the value of Y based on X

To understand how X affects Y

To identify the trend or relationship between two variables
















Q2. What are the key assumptions of Simple Linear Regression?
>- Key assumptions of Simple Linear Regression:

>Linearity ‚Äì A linear relationship exists between X and Y

>Independence ‚Äì Observations are independent of each other

>Homoscedasticity ‚Äì Constant variance of errors across all values of X

>Normality ‚Äì Errors are normally distributed

>No significant outliers ‚Äì No extreme values that strongly influence the model

Q5.  What is the method of least squares in linear regression?
>- The method of least squares is a technique used in linear regression to find the best-fit line by minimizing the sum of squared differences between the actual values and the predicted values.

Q6. What is Logistic Regression? How does it differ from Linear Regression?
>- Logistic Regression is a statistical model used for binary classification problems. It predicts the probability that an outcome belongs to a certain class (e.g., 0 or 1) using a sigmoid (logistic) function.

>- Difference from Linear Regression:

Output: Logistic regression predicts probabilities (0 to 1), while linear regression predicts continuous values.

Function: Logistic regression uses a sigmoid curve, linear regression uses a straight line.

Use case: Logistic regression is used for classification, linear regression for prediction/estimation.

Q7.  Name and briefly describe three common evaluation metrics for regression
models?
>- Three common evaluation metrics for regression models:

Mean Absolute Error (MAE) ‚Äì Average of absolute differences between actual and predicted values.

Mean Squared Error (MSE) ‚Äì Average of squared differences between actual and predicted values; penalizes large errors more.

R-squared (R¬≤) ‚Äì Measures how well the model explains the variance in the dependent variable.

Q8.  What is the purpose of the R-squared metric in regression analysis?
>- The purpose of R-squared (R¬≤) is to measure how well a regression model explains the variation in the dependent variable.
It indicates the goodness of fit of the model, with values closer to 1 showing better explanatory power.

Q9.  Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.
(Include your Python code and output in the code box below.)





```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
